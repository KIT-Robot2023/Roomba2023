```python
import matplotlib.pyplot as plt
import pandas
```

```python
df = pandas.read_csv('CSV_20240118203139.csv')
```

```python
x_t0 = []
x_t1 = []
x_t2 = []
y_t0 = []
y_t1 = []
y_t2 = []
for i in range(len(df[' x_t[m] '])):
    if df['target_point'][i] == 0:
        x_t0.append(df[' x_t[m] '][i])
    if df['target_point'][i] == 1:
        x_t1.append(df[' x_t[m] '][i])
    if df['target_point'][i] == 2:
        x_t2.append(df[' x_t[m] '][i])
x_ts = [x_t0, x_t1, x_t2]

for i in range(len(df[' y_t[m] '])):
    if df['target_point'][i] == 0:
        y_t0.append(df[' y_t[m] '][i])
    if df['target_point'][i] == 1:
        y_t1.append(df[' y_t[m] '][i])
    if df['target_point'][i] == 2:
        y_t2.append(df[' y_t[m] '][i])
y_ts = [y_t0, y_t1, y_t2]

theta0 = []
theta1 = []
theta2 = []
time0 = []
time1 = []
time2 = []
for i in range(len(df['theta[deg]'])):
    if df['target_point'][i] == 0:
        theta0.append(df['theta[deg]'][i])
        time0.append(df['time[s] '][i])
    if df['target_point'][i] == 1:
        theta1.append(df['theta[deg]'][i])
        time1.append(df['time[s] '][i])
    if df['target_point'][i] == 2:
        theta2.append(df['theta[deg]'][i])
        time2.append(df['time[s] '][i])
thetas = [theta0, theta1, theta2]
times = [time0, time1, time2]
target_point_x = (0.5, 0.5, 1.0)
target_point_y = (-0.5, 0, 0)
```

```python
plt.rcParams['xtick.direction'] = 'in'  # グラフの目盛の向き'in'で内向き
plt.rcParams['ytick.direction'] = 'in'  # グラフの目盛の向き'in'で内向き
plt.rcParams["font.family"] = "Cambria" # 使用するフォント
plt.rcParams["font.size"] = 10          # 文字の大きさ
```

```python
fig, ax = plt.subplots()
ax.scatter(y_ts[0], x_ts[0], label="Roomba's trajectory", facecolor='None', edgecolors='blue')
ax.scatter(y_ts[1], x_ts[1], label="Roomba's trajectory", facecolor='None', edgecolors='orange')
ax.scatter(y_ts[2], x_ts[2], label="Roomba's trajectory", facecolor='None', edgecolors='green')
ax.scatter(target_point_y, target_point_x, label='target point', marker='^', facecolor='None', edgecolors='red')
ax.set_xlabel ('Y[m]')
ax.set_ylabel('X[m]')
ax.set_xlim(-0.6, -0.6+1.2)
ax.set_ylim(-0.1, -0.1+1.2)
ax.legend(edgecolor='black',fancybox=False)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(times[0], thetas[0], label="Roomba's angele 1", facecolor='None', edgecolors='blue')
ax.scatter(times[1], thetas[1], label="Roomba's angele 2", facecolor='None', edgecolors='orange')
ax.scatter(times[2], thetas[2], label="Roomba's angele 3", facecolor='None', edgecolors='green')
ax.set_xlabel ('time[s]')
ax.set_ylabel('angle[°]')
ax.legend(edgecolor='black',fancybox=False)
plt.show()
```
